```python
def plot_confusion_matrix(cm, classes,
                        normalize=False,
                        title='Confusion matrix',
                        cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
            horizontalalignment="center",
            color="white" if cm[i, j] > thresh else "black")

    #plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

# we want TNR to go higher and FNR to go lower
def True_Negative_rate(cm_obj):
    return(cm_obj[0][0]/(cm_obj[0][0] + cm_obj[0][1])) 

def False_Negative_rate(cm_obj):
    return(cm_obj[1][0]/(cm_obj[1][0] + cm_obj[1][1]))

#metric_tnr = make_scorer(tnr, greater_is_better=True) # custom score function 
def result(X,y,obj,normalize = True):
    cm = confusion_matrix(y, obj.predict(X))
    print(f" TN rate is {True_Negative_rate(cm)}")
    print(f" FN rate is {False_Negative_rate(cm)}")
    print(f" TP rate is {1-False_Negative_rate(cm)}")
    print(f" FP rate is {1-True_Negative_rate(cm)}")
    
    plot_confusion_matrix(cm,classes=['Pass','Fail'],title="Confusion Matrix",normalize=False)
```

#### Light GBM Fine tuning using Optuna

```python
def lgb_objective(trial,x_train,y_train):
       # 2. Suggest values for the hyperparameters using a trial object.
        val_dict = {-1:len(y_train)/(2*Counter(y_train)[-1]),1:len(y_train)/(2*Counter(y_train)[1])}

        classifier_name = trial.suggest_categorical('classifier', ['LGBM'])
    
        param = {
        'objective': 'binary',
        'metric': 'binary_logloss',
        'verbosity': -1,
        'boosting_type': 'gbdt',
        'lambda_l1': trial.suggest_loguniform('lambda_l1', 1e-8, 10.0),
        'lambda_l2': trial.suggest_loguniform('lambda_l2', 1e-8, 10.0),
        'num_leaves': trial.suggest_int('num_leaves', 2, 256),
        'feature_fraction': trial.suggest_uniform('feature_fraction', 0.1, 1.0),
        'bagging_fraction': trial.suggest_uniform('bagging_fraction', 0.1, 1.0),
        'bagging_freq': trial.suggest_int('bagging_freq', 1, 7),
        'min_child_samples': trial.suggest_int('min_child_samples', 5, 100),
        'class_weight':val_dict,
        'n_estimators':trial.suggest_int('rf_n_estimators',1,3000),
        'learning_rate':trial.suggest_loguniform('rf_lr',1e-5,1e2),
        'max_depth':int(trial.suggest_loguniform('rf_max_depth', 2, 9))
        
                        }
        
        classifier_obj =lgbm.LGBMClassifier(**param)
        score = sklearn.model_selection.cross_val_score(
            classifier_obj,
            x_train,
            y_train,
            n_jobs=-1,
            cv=2,
            scoring = 'roc_auc'
        )
        accuracy = score.mean()
        return accuracy
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

def fit_lgbm(X, y, params=None, es_rounds=20, seed=42, N_SPLITS=5, 
             n_class=None, model_dir=None, folds=None):
    models = []
    lda_model = []
    oof = np.zeros((len(y), n_class), dtype=np.float64)
    
    for i in tqdm(range(CFG.n_splits)):
        print(f"== fold {i} ==")
        trn_idx = folds!=i
        val_idx = folds==i
        X_train, y_train = X[trn_idx], y.iloc[trn_idx]
        X_valid, y_valid = X.iloc[val_idx], y.iloc[val_idx]
        
        lda = LinearDiscriminantAnalysis()
        lda.fit(X_train,y_train)
        X_train['lda_feat'] = lda.transform(X_train)
        X_valid['lda_feat'] = lda.transform(X_valid)
        lda_model.append(lda)
        
        
        if model_dir is None:
            lgb_study = optuna.create_study(direction="maximize")
            func = lambda trial : lgb_objective(trial,X_train,y_train)
            lgb_study.optimize(func, n_trials=20)
            val_dict = {-1:len(y_train)/(2*Counter(y_train)[-1]),1:len(y_train)/(2*Counter(y_train)[1])}
            params = {
                'objective': 'binary',
                'metric': 'binary_logloss',
                'verbosity': -1,
                'boosting_type': 'gbdt',
                'lambda_l1': lgb_study.best_params['lambda_l1'],
                'lambda_l2': lgb_study.best_params['lambda_l2'],
                'num_leaves': lgb_study.best_params['num_leaves'],
                'feature_fraction': lgb_study.best_params['feature_fraction'],
                'bagging_fraction': lgb_study.best_params['bagging_fraction'],
                'bagging_freq': lgb_study.best_params['bagging_freq'],
                'min_child_samples': lgb_study.best_params['min_child_samples'],
                 'class_weight':val_dict,
                'n_estimators':lgb_study.best_params['rf_n_estimators'],
                'learning_rate':lgb_study.best_params['rf_lr'],
                'max_depth':int(lgb_study.best_params['rf_max_depth'])
            }

            
            model = lgbm.LGBMClassifier(**params)

            model.fit(X_train,y_train)
            
        else:
            with open(f'{model_dir}/lgbm_fold{i}.pkl', 'rb') as f:
                model = pickle.load(f)
            
        pred = model.predict_proba(X_valid)
        oof[val_idx] = pred
        models.append(model)
        
        file = f'lgbm{i}.pkl'
        pickle.dump(model, open(file, 'wb'))

    cv = (oof.argmax(axis=-1) == y).mean()
    print(f"CV-accuracy: {cv}")

    return oof, models,lda_model
```

#### XGBoost Fine tuning using Optuna

```python
def xgb_objective(trial,x_train,y_train):
    
        ratio = float(np.sum(y_train == 1)) / np.sum(y_train==-1)

        param = {
            'reg_alpha': trial.suggest_loguniform('reg_alpha', 1e-3, 10.0),
            'subsample': trial.suggest_uniform('subsample', 0.3, 1.0),
            'colsample_bytree': trial.suggest_uniform('colsample_bytree', 0.4, 1.0),
            'n_estimators':trial.suggest_int('rf_n_estimators',1,2000),
            'learning_rate':trial.suggest_loguniform('rf_lr',1e-5,1e2),
            'max_depth':int(trial.suggest_loguniform('rf_max_depth', 2, 7)),
            'scale_pos_weight':ratio

        }



        clf = xgb.XGBClassifier(**param)

        score = sklearn.model_selection.cross_val_score(
                clf,
                x_train,
                y_train,
                n_jobs=-1,
                cv=2,
                scoring = 'roc_auc'
            )

        accuracy = score.mean()
        return accuracy

#     clf.fit(X_train, y_train, early_stopping_rounds=50, eval_metric="auc",
#             eval_set=[(X_train, y_train), (X_valid, y_valid)])
```

```python
def fit_xgb(X, y, params=None, es_rounds=20, seed=42, N_SPLITS=5, 
             n_class=None, model_dir=None, folds=None):
    models = []
    oof = np.zeros((len(y), n_class), dtype=np.float64)
    
    for i in tqdm(range(CFG.n_splits)):
        print(f"== fold {i} ==")
        trn_idx = folds!=i
        val_idx = folds==i
        X_train, y_train = X[trn_idx], y.iloc[trn_idx]
        X_valid, y_valid = X.iloc[val_idx], y.iloc[val_idx]
        
#         X_train,X_valid = select_features(X_train,y_train,X_valid,num_feat=5)
        
        if model_dir is None:
            xgb_study = optuna.create_study(direction="maximize")
            func = lambda trial : xgb_objective(trial,X_train,y_train)
            xgb_study.optimize(func, n_trials=20)
            ratio = float(np.sum(y_train == 1)) / np.sum(y_train==-1)
            params = {
                    'reg_alpha': xgb_study.best_params['reg_alpha'],
                    'subsample': xgb_study.best_params['subsample'],
                    'colsample_bytree': xgb_study.best_params['colsample_bytree'],
                    'n_estimators':xgb_study.best_params['rf_n_estimators'],
                    'learning_rate':xgb_study.best_params['rf_lr'],
                    'max_depth':int(xgb_study.best_params['rf_max_depth']),
                    'scale_pos_weight':ratio}

            
            model = xgb.XGBClassifier(**params)

            model.fit(X_train,y_train)
            
        else:
            with open(f'{model_dir}/xgb_fold{i}.pkl', 'rb') as f:
                model = pickle.load(f)
            
        pred = model.predict_proba(X_valid)
        oof[val_idx] = pred
        models.append(model)
        
        file = f'xgb{i}.pkl'
        pickle.dump(model, open(file, 'wb'))

    cv = (oof.argmax(axis=-1) == y).mean()
    print(f"CV-accuracy: {cv}")

    return oof, models

```
